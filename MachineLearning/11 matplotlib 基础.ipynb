## 11 matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.show() # no need with this line the previous line can draw this pic
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
# plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="pink")
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, linestyle="-.")
```

```python
mpl.__version__
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, linestyle="-.")
plt.xlim(-5,15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, linestyle="-.")
plt.axis([-1, 11, -2, 2])
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, linestyle="-.")
plt.xlabel("x axis")
plt.ylabel("y value")
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, linestyle="-.", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, linestyle="-.", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("welcome to the  ML world!")
```

### Scatter Plot

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="pink")
plt.show()
```
